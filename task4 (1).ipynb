```python
d = {'k1' :"value" , "k2" : "values " ,"k3" : { "k12" : "sudh" , "k13"  : "gafasd"}}
```

```python
def dic_to_list(d) : 
    l= []
    for i in d.values() : 
        if type(i) != dict :
            l.append(i)
        if type(i) == dict :
            for j in i.values():
                l.append(j)
    return l 
```

```python
dic_to_list(d)
```
